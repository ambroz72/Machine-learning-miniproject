```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('BitcoinPrice.csv')
```

```python
df.drop(labels=['Date'], axis=1, inplace=True)
```

```python
df.head()
```

```python
p_days = 30
df['Prediction']=df[['Price']].shift(-p_days)
```

```python
df.head()
```

```python
df.tail()
```

```python
x=np.array(df.drop(['Prediction'], axis=1,))
x= x[: len(df) - p_days]
```

```python
x.shape
```

```python
x
```

```python
y = np.array(df['Prediction'])
y= y[: -p_days]
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
P_days_array = np.array(df.drop(['Prediction'], axis=1))[-p_days: ]
```

```python
print(P_days_array)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf= RandomForestRegressor(n_estimators = 1000, random_state = 1)
Rf.fit(x_train, y_train)
print('Random Forcast Accuracy: {:.2f}%'.format(Rf.score(x_test, y_test)*100))
```

```python
Rf_predict = Rf.predict(x_test)
print(Rf_predict)
```

```python
Rf_predict_30= Rf.predict(P_days_array)
print(Rf_predict)
```
